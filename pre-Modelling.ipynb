## text Preprocessing

### Case Folding

```python
df["Comment"] = df["Comment"].str.lower()
df["Username"] = df["Username"].str.lower()
```

```python
df
```

### Penghapusan tanda baca

```python
import string
def hapus_tanda_baca(teks):
    return ''.join(karakter for karakter in teks if karakter not in string.punctuation)
df["Comment"] = df["Comment"].apply(hapus_tanda_baca)
```

```python
df
```

### Penghapusan karakter khusus / html

```python
from bs4 import BeautifulSoup
import re
def remove_html(teks):
    soup = BeautifulSoup(teks, "html.parser")
    return soup.get_text()
df["Comment"] = df["Comment"].apply(remove_html)
```

```python
df
```

### penghapusan karakter khusus ! # @ dll

```python
df["Comment"] = df["Comment"].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df["Username"] = df["Username"].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

```python
df
```

### menangani whitespace

```python
df["Comment"] = df["Comment"].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df
```

## Normalisasi

### Tokenization

```python
from nltk.tokenize import word_tokenize
df["Comment"] = df["Comment"].apply(word_tokenize)
```

```python
df
```

### Normalisasi kata

```python
# URL file kamus slang
url_slang = 'https://raw.githubusercontent.com/nasalsabila/kamus-alay/master/colloquial-indonesian-lexicon.csv'

df_slang = pd.read_csv(url_slang)
print(df_slang.head())
# change to dictionery
slang_dict = dict(zip(df_slang['slang'], df_slang['formal']))
```

```python
def normalisasi_slang(teks):
    normalisasi_token = []
    for word in teks:
        normalisasi_token.append(slang_dict.get(word, word))
    return normalisasi_token
```

```python
df["Comment"] = df["Comment"].apply(normalisasi_slang)
```

```python
df
```

### penghapusan Stopword

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stop_word_id = factory.get_stop_words()

def remove_stopword(teks):
    filter_tokens = [word for word in teks if word not in stop_word_id]
    return filter_tokens
```

```python
df["Comment"] = df["Comment"].apply(remove_stopword)
```

```python
df
```

### Stemming

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer_id = factory.create_stemmer()

def sastrawi_stemming_kata(kata_tunggal):
    return stemmer_id.stem(kata_tunggal)

def proses_stemming_per_komentar(list_kata_komentar):
    if isinstance(list_kata_komentar, list):
        komentar_stemmed = [sastrawi_stemming_kata(kata) for kata in list_kata_komentar]
        return komentar_stemmed
    else:
        return sastrawi_stemming_kata(str(list_kata_komentar))
df["Comment"] = df["Comment"].apply(proses_stemming_per_komentar)
```
